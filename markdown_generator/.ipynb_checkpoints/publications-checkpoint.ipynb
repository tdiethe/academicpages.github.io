# Publications markdown generator for academicpages

Takes the .bib files of publications converts them for use with [academicpages.github.io](academicpages.github.io). This is an interactive Jupyter notebook ([see more info here](http://jupyter-notebook-beginner-guide.readthedocs.io/en/latest/what_is_jupyter.html)). The core python code is also in `publications.py`. Run either from the `markdown_generator` folder.


## Data format

The TSV needs to have the following columns: pub_date, title, venue, excerpt, citation, site_url, and paper_url, with a header at the top. 

- `excerpt` and `paper_url` can be blank, but the others must have values. 
- `pub_date` must be formatted as YYYY-MM-DD.
- `url_slug` will be the descriptive part of the .md file and the permalink URL for the page about the paper. The .md file will be `YYYY-MM-DD-[url_slug].md` and the permalink will be `https://[yourdomain]/publications/YYYY-MM-DD-[url_slug]`

This is how the raw file looks (it doesn't look pretty, use a spreadsheet or other program to edit and create).

````python
from pybtex.database import parse_file
from pybtex.database.input import bibtex
# from pybtex.database.output import yaml
import pybtex

print(pybtex.__version__)

parser = bibtex.Parser()
bib_data = parser.parse_file('../_bibliography/main.bib')
bib_data.to_string(bib_format='yaml')
print(bib_data.entries.values()[0])
````

````python
import bibtexparser
from bibtexparser.bparser import BibTexParser
from bibtexparser.customization import convert_to_unicode, homogeneize_latex_encoding
from bibtexparser.bibdatabase import BibDatabase
import sys

with open('../_bibliography/main.bib', encoding=sys.getdefaultencoding()) as bibtex_file:
    bibtex_str = bibtex_file.read()

def add_plaintext_fields(record):
    """
    For each field in the record, add a `plain_` field containing the
    plaintext, stripped from braces and similar. See
    https://github.com/sciunto-org/python-bibtexparser/issues/116.

    :param record: the record.
    :type record: dict
    :returns: dict -- the modified record.
    """
    def _strip_string(string):
        return string.replace("{", "").replace("}", "")

    modified = dict()
    for key in record.keys():
        if isinstance(record[key], str):
            modified[key] = _strip_string(record[key])
        elif isinstance(record[key], dict):
            modified[key] = {subkey: _strip_string(value) for subkey, value in record[key].items()}
        elif isinstance(record[key], list):
            modified[key] = list(map(_strip_string, record[key]))
        else:
            raise ValueError(record[key])

    if record["ID"] == "Diethe07":
        print("Record")
        print(record)
        print("modified")
        print(modified)
        print()
    return modified
    
def dump_single(entry):
    db = BibDatabase()
    db.entries = [entry]
    return bibtexparser.dumps(db)

parser = BibTexParser()
parser.customization = add_plaintext_fields  # lambda x: homogeneize_latex_encoding(add_plaintext_fields(x))
bib_database = bibtexparser.loads(bibtex_str, parser=parser)
parser = BibTexParser()
parser.customization = lambda x: add_plaintext_fields(convert_to_unicode(x))
bib_unicode = bibtexparser.loads(bibtex_str, parser=parser)

print(bib_database.entries[-10])
print()
print(bib_unicode.entries[-10])
print()
print(dump_single(bib_database.entries[-10]))
````

## Import TSV

Pandas makes this easy with the read_csv function. We are using a TSV, so we specify the separator as a tab, or `\t`.

I found it important to put this data in a tab-separated values format, because there are a lot of commas in this kind of data and comma-separated values can get messed up. However, you can modify the import statement, as pandas also has read_excel(), read_json(), and others.

````python
publications = pd.read_csv("publications.tsv", sep="\t", header=0)
publications
````

````python
publications = pd.DataFrame(bib_database.entries)
publications["citation"] = pd.Series(map(dump_single, bib_database.entries))
publications["citation_unicode"] = pd.Series(map(dump_single, bib_unicode.entries))
publications = publications[publications.year.apply(lambda x: x.isnumeric())]  # strip out things like "In preparation"
del publications["date-added"]
del publications["date-modified"]
del publications["owner"]
del publications["editor_removed"]
del publications["authorshort"]
publications["pub_date"] = publications.year.str.replace(r'(\d+)', r'\1-10-01')
publications["url_slug"] = publications.ID.str.lower()
publications["excerpt"] = publications.abstract
publications["paper_url"] = ""

def selector(x):
    if x.ENTRYTYPE == "incollection":
        return x.booktitle
    if x.ENTRYTYPE == "book":
        return x.publisher
    if x.ENTRYTYPE == "inproceedings":
        return x.booktitle
    if x.ENTRYTYPE == "techreport":
        return x.institution
    if "thesis" in x.ENTRYTYPE:
        return x.school
    if x.ENTRYTYPE == "article":
        return x.journal
    raise ValueError(x.ENTRYTYPE)
    
publications["venue"] = publications.apply(selector, axis=1)
# publications[publications.ENTRYTYPE == "techreport"].head()
publications.head()
````

## Escape special characters

YAML is very picky about how it takes a valid string, so we are replacing single and double quotes (and ampersands) with their HTML encoded equivilents. This makes them look not so readable in raw format, but they are parsed and rendered nicely.

````python
html_escape_table = {
    "&": "&amp;",
    '"': "&quot;",
    "'": "&apos;",
    '\"{o}': "ö",
    '\"o': "ö",
#     "{": "",
#     "}": ""
    }

def html_escape(text):
    """Produce entities within text."""
    return "".join(html_escape_table.get(c,c) for c in text)
````

## Creating the markdown files

This is where the heavy lifting is done. This loops through all the rows in the TSV dataframe, then starts to concatentate a big string (```md```) that contains the markdown for each type. It does the YAML metadata first, then does the description for the individual page.

````python
import os
for row, item in publications.iterrows():
    
    md_filename = str(item.pub_date) + "-" + item.url_slug + ".md"
    html_filename = str(item.pub_date) + "-" + item.url_slug
    year = item.pub_date[:4]
    
    ## YAML variables
    
    md = "---\ntitle: \""   + html_escape(item.title) + '"\n'
    
    md += """collection: publications"""
    
    md += """\npermalink: /publication/""" + html_filename
    
    if len(str(item.excerpt)) > 5:
        md += "\nexcerpt: '" + html_escape(item.excerpt) + "'"
    
    md += "\ndate: " + str(item.pub_date) 
    
    md += "\nvenue: '" + html_escape(item.venue) + "'"
    
    if len(str(item.paper_url)) > 5:
        md += "\npaperurl: '" + item.paper_url + "'"
    
    md += "\ncitation: '" + html_escape(item.citation) + "'"
    
#     md += "\n---"
    
    ## Markdown description for individual page
        
#     if len(str(item.excerpt)) > 5:
#         md += "\n" + html_escape(item.excerpt) + "\n"
    
#     if len(str(item.paper_url)) > 5:
#         md += "\n[Download paper here](" + item.paper_url + ")\n" 
        
    #     md += "\nRecommended citation: " + item.citation_unicode
    
    md_filename = os.path.basename(md_filename)
       
    with open("../_publications/" + md_filename, 'w', encoding=sys.getdefaultencoding()) as f:
        f.write(md)
````
